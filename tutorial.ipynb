```python
import os
from getpass import getpass

import requests

# prepare the OpenAI API key
os.environ['OPENAI_API_KEY'] = getpass("Enter your OpenAI API key: ")

# prepare the sample PDF
response = requests.get(
    "https://github.com/luisleo526/doc2mark/raw/492e968e59eae4744fee3290c75b509f7fa30eb7/sample_documents/complex-tables/complex_table_test.pdf"
)
with open("complex_table_test.pdf", "wb") as f:
    f.write(response.content)
```

```python
from doc2mark import UnifiedDocumentLoader
from IPython.display import display, HTML

# Basic usage
loader = UnifiedDocumentLoader()

result = loader.load(
    "complex_table_test.pdf",
    extract_images=True,
    ocr_images=True
)

# Display the result
# If result.content is a string containing HTML/Markdown
display(HTML(result.content))
```

```python
from doc2mark import OCRConfig

# change to other language
loader = UnifiedDocumentLoader(
    ocr_config=OCRConfig(
        language="Japanese"
    )
)

result = loader.load(
    "complex_table_test.pdf",
    extract_images=True,
    ocr_images=True
)

# Display the result
# If result.content is a string containing HTML/Markdown
display(HTML(result.content))
```
